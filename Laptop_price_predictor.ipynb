```python
def add_company(input):
  if input=='Samsung' or input=='Razer' or input=='mediacom' or input=='Microsoft' or input=='Xiaomi' or input=='Vero' or input=='Chuwi' or input=='Google' or input=='Fujitsu' or input=='LG' or input=='Huawei':
    return 'Other'
  else:
    return input

data['Company']=data['Company'].apply(add_company)
```

```python
data['Touchscreen']=data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips']=data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
data.head(2)
```

```python
data['Cpu'].value_counts()
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['cpu_name'].value_counts()
```

```python
def set_processor(name):
  if name=='intel Core i7' or name=='intel Core i5' or name=='intel Core i3':
    return name
  else:
    if name.split()[0] == 'AMD':
      return 'AMD'
    else:
      return 'Other'

data['cpu_name']=data['cpu_name'].apply(set_processor)
```

```python
data['cpu_name'].value_counts()
```

```python
data['Gpu'].value_counts()
```

```python
len(data['Gpu'].value_counts())
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
data['gpu_name'].value_counts()
```

```python
data.shape
```

```python
data= data[data['gpu_name'] != 'ARM']
```

```python
data.shape
```

```python
data.head(5)
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(i):
  if i== 'Windows 10' or i=='Windows 7' or i=='Windows 10 S':
    return 'Windows'
  elif i=='macOS' or i=='Mac OS X':
    return 'Mac'
  elif i=='Linux':
    return 'Linux'
  else:
    return 'Other'

data['OpSys']=data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.head(2)
```

```python
data=data.drop(columns=['Inches','Product'])
```

```python
data.head(2)
```

```python
data=pd.get_dummies(data)
```

```python
data.head(2)
```

```python
data.shape
```

```python
x=data.drop('Price_euros',axis=1)
y=data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
```

```python
x_train.shape,x_test.shape
```

```python
def model_acc(model):
  model.fit(x_train,y_train)
  acc=model.score(x_test,y_test)
  print(str(model)+'-->'+str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso=Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
model_acc(rf)

```

```python
from sklearn.model_selection import GridSearchCV

parameters={'n_estimators':[10,50,100],'criterion':['squared_error','absolute_error', 'poisson']}


grid_obj=GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit=grid_obj.fit(x_train,y_train)

best_model=grid_fit.best_estimator_

best_model


```

```python
best_model.score(x_test,y_test)
```

```python
best_model.score(x_train,y_train)
```

```python
import pickle
with open('predictor.pickle','wb') as file:
  pickle.dump(best_model,file)
```
